```python
import sys, pygame
import random as rnd
pygame.init()

size = width, height = 640, 480
speed = [0, 0]
black = 0, 0, 0

screen = pygame.display.set_mode(size)

duck = pygame.image.load("duck_PNG5023.png")
food = pygame.image.load("685-6850444_coral-algae-seaweed-plant-algae-with-transparent-background.png")
duckrect = duck.get_rect()
foodrect = food.get_rect()
pygame.font.init() 
myfont = pygame.font.SysFont('Comic Sans MS', 30)
count = 0
# #         if count % 5 == 0:
x_coord = rnd.randint(0 + 60, width-60)
y_coord = rnd.randint(0 + 60, height-60)
# foodrect.left = x_coord
# #             screen.fill('blue')
# #             screen.blit(ball, ballrect)
screen.blit(food, (x_coord, y_coord))
pygame.time.delay(10000000) 
check_food = False
while 1:
    for event in pygame.event.get():
        count += 1
#         x_coord = rnd.randint(0 + 60, width-60)
#         y_coord = rnd.randint(0 + 60, height-60)        
        if count % 10 == 0:
            check_food = True
            x_coord = rnd.randint(0 + 60, width-60)
            y_coord = rnd.randint(0 + 60, height-60)    
            screen.blit(food, (x_coord, y_coord))
            pygame.display.update() 
            pygame.time.delay(20000)
            
        if event.type == pygame.QUIT: 
            pygame.quit()
            sys.exit()
            
        if event.type == pygame.MOUSEBUTTONDOWN: 
            pos = pygame.mouse.get_pos()
#             print(pos)
            textsurface = myfont.render('КРЯ', True, (50,50,50))
            screen.blit(textsurface,pos)
            print(duckrect.top, duckrect.bottom, y_coord)
#             textsurface = myfont.render('КРЯ', False, (50,50,50))
#             screen.blit(textsurface,pos)

#             myfont.render_to(screen, pos, "Hello World!", (255, 56, 255))
#             pygame.display.flip()

#             print('КРЯ')
            
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP:
                speed[1] -= 2
            if event.key == pygame.K_DOWN:
                speed[1] += 2
            if event.key == pygame.K_LEFT:
                speed[0] -= 2
            if event.key == pygame.K_RIGHT:
                speed[0] += 2
            if event.key == pygame.K_SPACE:
                speed[1] -= 10
            if event.key == pygame.K_RIGHT and event.key == pygame.K_SPACE:
                        speed[1] -= 10
                        speed[0] += 2
                        
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_UP:
                speed[1] = 0
            if event.key == pygame.K_DOWN:
                speed[1] = 0
            if event.key == pygame.K_LEFT:
                speed[0] = 0
            if event.key == pygame.K_RIGHT:
                speed[0] = 0
            if event.key == pygame.K_SPACE:
                speed[1] += 10
        
        if check_food and y_coord*0.5 < duckrect.top < y_coord*1.5 and x_coord*0.5 < duckrect.right < x_coord*1.5:
            print('eat')
            screen.fill('blue')
            screen.blit(duck, duckrect)
#             foodrect.update(0, 0,0,0)
            screen.blit(food, (600, 400))

#             foodrect.fill((0, 0, 0, 0))
            textsurface = myfont.render('НЯММММММММММММ', True, (50,50,50))
            screen.blit(textsurface, (x_coord, y_coord))
            pygame.time.delay(200)
            pygame.display.update() 

            pygame.display.flip()


            
#                 if event.type == pygame.KEYUP:

    duckrect = duckrect.move(speed)

    if duckrect.left < 0 or duckrect.right > width:
        #speed[0] = -speed[0]
        speed[0] = 0
        
    if duckrect.top < 0 or duckrect.bottom > height:
        #speed[1] = -speed[1]
        speed[1] = 0
        
    screen.fill('blue')
    screen.blit(duck, duckrect)
    screen.blit(food, (x_coord, y_coord))
#     print((x_coord, y_coord))

    pygame.display.flip()
    
    pygame.time.delay(20)
```
